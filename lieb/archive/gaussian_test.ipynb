```python
import numpy as np

import plotly.graph_objects as go
```

$$V_+(x)=Ae^{-\frac{x^2}{2\sigma^2}}$$
$$V_+(k)=\sqrt{2\pi}\sigma Ae^{-\pi^2 2\sigma^2k^2}$$


```python
n_k = 10000

k = np.linspace(-1/2, 1/2, n_k)
x = np.fft.fftshift(np.fft.fftfreq(n=n_k, d=1/n_k))
```

```python
A = 2
sigma = 100

V_k = A * np.sqrt(2*np.pi) * sigma * np.exp(-np.pi**2 * 2*sigma**2 * (k)**2)

V_x_theory = A*np.exp(-x**2/(2*sigma**2))
V_x_fft = np.real(np.fft.fftshift(np.fft.fft(np.fft.ifftshift(V_k)))) / n_k
```

```python
figure = go.Figure(data=[
    go.Scatter(x=k, y=V_k, name="Position"), 
])
figure.update_layout(autosize=False, width=750, height=750)
figure.show()
```

```python
figure = go.Figure(data=[
    go.Scatter(x=x, y=V_x_theory, name="Theory"), 
    go.Scatter(x=x, y=V_x_fft, name="FFT"),
])
figure.update_layout(autosize=False, width=750, height=750)
figure.show()
```
