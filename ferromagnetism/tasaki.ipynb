```python
import numpy as np

import plotly.graph_objects as go
```

```python
N_s = 2000
t = 1
```

```python
T = np.zeros((N_s, N_s))

for x in range(N_s):
    T[x, (x+1)%N_s] = -t
    T[(x+1)%N_s, x] = -t
```

```python
e, eta = np.linalg.eigh(T)
```

```python
x = np.arange(0, N_s)
n = np.arange(-N_s//2, N_s//2)
k = 2*np.pi/N_s * n
```

```python
figure = go.Figure(data=[go.Scatter(x=k, y=e)])
figure.show()
```

```python
figure = go.Figure(data=[go.Scatter(x=x, y=eta[:, 1])])
figure.show()
```

```python
figure = go.Figure(data=[go.Scatter(x=x, y=np.abs(eta[:, 1])**2)])
figure.show()
```
